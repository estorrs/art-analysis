```python
import subprocess

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import skimage
from skimage.color import rgb2hsv, rgba2rgb, hsv2rgb, rgb2gray
```

```python
%matplotlib notebook
```

```python
fp = '../data/portrait/th (38).jpeg'
root = fp.split('/')[-1].split('.')[0]
root
```

```python
img = skimage.io.imread(fp)

if img.shape[2] == 4:
    img = rgba2rgb(img)
    
if np.max(img) > 1:
    img = img / 255.

hsv_img = rgb2hsv(img)
gray_img = rgb2gray(img)
plt.imshow(img)
plt.show()
```

###### sandbox

```python
def on_pick(event):
    x, y = event.xdata, event.ydata
    
    ax1.scatter(x, y, s=2, color='black')
    ax3.scatter(x, y, s=2, color='#ab5252')
    
    r, g, b = img[int(y), int(x)]
    ax5.axvline(gray_img[int(y), int(x)] * 256., -.5, 2.5, color=(r, g, b))
    
    h, s, v = hsv_img[int(y), int(x)]
    
    p = 0.01745329 * (h * 360)
    
    x, y = s * np.cos(p), s * np.sin(p)
    
    ax2.scatter(x, y, v, color=(r, g, b))
    ax4.scatter(x, y, color=(r, g, b))
    
    fig.canvas.draw()
    
def add_hues(ax):
    for i in range(0, 360, 5):
        h, s, v = i / 360., 1., 1.
        r, g, b = hsv2rgb(np.asarray([h, s, v]).reshape((1, 1, -1))).flatten()

        p = 0.01745329 * (h * 360)
        x, y = s * np.cos(p), s * np.sin(p)

        ax.scatter(x, y, color=(r, g, b))
```

```python
fig = plt.figure(figsize=(10, 15))
ax1 = fig.add_subplot(3, 2, 1)
im = ax1.imshow(img)

ax2 = fig.add_subplot(3, 2, 2, projection='3d')
ax2.set_xlim(-1., 1.)
ax2.set_ylim(-1., 1.)
ax2.set_zlim(-1., 1.)

ax3 = fig.add_subplot(3, 2, 3,)
im = ax3.imshow(gray_img, cmap='Greys_r')

ax4 = fig.add_subplot(3, 2, 4)
ax4.set_xlim(-1.1, 1.1)
ax4.set_ylim(-1.1, 1.1)

ax5 = fig.add_subplot(3, 2, 5)
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))
ax5.set_ylim(-.5, 2.5)
ax5.imshow(gradient, aspect='auto', cmap='Greys_r')

add_hues(ax4)

fig.canvas.mpl_connect('button_press_event', on_pick)
plt.show()
```
